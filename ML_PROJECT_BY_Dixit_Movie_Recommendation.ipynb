```python
def context(obj):
    L=[]              # Step 1: Create an empty list named 'L'

    for i in ast.literal_eval(obj):# Step 2: Convert 'obj' (a string) into a Python object (e.g., a list or dict) and loop through it

        L.append(i['name'])# Step 3: Access the 'name' key of each item in the loop and add its value to the list 'L'
    return L # Step 4: Return the list 'L'
```

```python
def context_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<3:
            L.append(i["name"])
            counter +=1
        else:
            break
    return L
```

```python
Movies['cast']=Movies['cast'].apply(context_cast)
```

```python
Movies.head()
```

```python
# now the last column remaining is crew. crew are which works behind the cameras like(directors,producers,cameramne etc)
# so in crew i think what important is director
# i will search like suggest me some movies of christopher nolan etc
# most of the time we remember directors name not cameramnen names so.
```

```python
Movies.iloc[1].crew
```

```python
def context_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
        
```

```python
Movies['crew']=Movies['crew'].apply(context_crew)
```

```python
# now next thing we will do is remove space from name in genres,keywords,cast and crew.
```

```python
# so let's write a program to remove space between words
```

```python
def collaspe(obj):
    L=[]
    for i in obj:
        L.append(i.replace(" ",''))
    return L
```

```python
Movies['crew']=Movies['crew'].apply(collaspe)
Movies['genres']=Movies['genres'].apply(collaspe)
Movies['keywords']=Movies['keywords'].apply(collaspe)
Movies['cast']=Movies['cast'].apply(collaspe)
```

```python
def recommend(movie):
    index= new_movie_data[new_movie_data['title']==movie].index[0]
    distances= sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    for i in distances[1:6]:
        print(new_movie_data.iloc[i[0]].title)
#         pass
```
